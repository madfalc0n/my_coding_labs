```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import cv2

import skimage.io
from skimage import color
```

```python
PATH = 'dataset/'
DATA_PATH = 'dataset/train_images/'
MASK_DATA_PATH = 'dataset/train_masks/'
SEED = 2020
```

## Data Load 

```python
train = pd.read_csv(PATH + 'train.csv')
# display(train.head(1))

# ISUP 3 이상인 애들만 추출
up_train = train[train['isup_grade'] >= 3]
up_train.reset_index(inplace=True,drop=True)
```

```python
#이미지 번호
image_index = 0

#이미지 호출위한 변수 설정
# image_file = DATA_PATH + up_train.iloc[image_index]['image_id'] + '.tiff'
# mask_image_file = MASK_DATA_PATH + up_train.iloc[image_index]['image_id'] + '_mask.tiff'

tmp_iamge = '00928370e2dfeb8a507667ef1d4efcbb'
image_file = DATA_PATH + tmp_iamge + '.tiff'
mask_image_file = MASK_DATA_PATH + tmp_iamge + '_mask.tiff'


#원본이미지 호출
image = skimage.io.MultiImage(image_file)[1]

#마스크이미지 호출, R채널에만 이미지 값이 있음
mask_image = skimage.io.MultiImage(mask_image_file)[1]
mask_image = mask_image[:,:,0]
mask_cnt_list = list(np.unique(mask_image))
mask_label = up_train['isup_grade'][image_index]
image_provider = up_train['data_provider'][image_index]

# Mask gleson score 값
#print(f'mask unique : {np.unique(mask_image)}')

# image show
fig, ax = plt.subplots(1, 2, figsize=(10, 7))
fig.suptitle(f"Provider : {image_provider}, ISUP_Grade : {mask_label}", fontsize=16)
ax[0].imshow(image)
ax[0].set_title(f'Original Image')
#cmap = matplotlib.colors.ListedColormap(['black', 'gray', 'green', 'yellow', 'orange', 'red'])
#ax[1].imshow(np.asarray(mask_image*20),cmap=cmap, interpolation='nearest', vmin=0, vmax=5)
ax[1].imshow(mask_image*20)
ax[1].set_title(f'Mask Image')
plt.show()

# Gleason score 별로 show
fig, ax = plt.subplots(1, len(mask_cnt_list), figsize=(15, 7))
fig.suptitle(f"Show Gleason Score", fontsize=16)
for i,value in enumerate(mask_cnt_list):
    cpy_mask_image = mask_image.copy()
    thresholds = (cpy_mask_image[:,:] == value)
    ax[i].imshow(thresholds)
    ax[i].set_title(f'Gleason Score : {value}')
plt.show()
```
